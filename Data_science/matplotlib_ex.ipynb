```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
print(plt.style.available)
plt.style.use('dark_background')
```

```python
plt.figure(figsize=(20,10))
t=np.linspace(0,4*np.pi,100)
y=np.sin(t)
x=np.cos(t)
#z=np.tan(t)
plt.plot(t,y, label='sin(t)',color='red',linewidth=3,marker='o',markersize=10)
plt.plot(t,x, label='cos(t)',color='white',linewidth=2,linestyle='-.')
plt.text(10,0.5,'sin(t)',fontsize=20)
plt.text(7,-0.5,'cos(t)',fontsize=20)

plt.legend()
#plt.plot(t,z)
plt.grid(False)
plt.savefig('sin_cos.png')
```

```python
plt.figure(figsize=(30,10))
s=np.linspace(0,4*np.pi,100)
x=np.tan(s)
y=np.cos(s)

plt.plot(s,x,label='tan(s)',color='yellow',linewidth=2,marker='D',markersize=10)
plt.plot(s,y,label='cos(s)',color='hotpink',linewidth=2,linestyle='--')
plt.text(10,0.5,'tan(s)',fontsize=20)
plt.text(7,-0.5,'cos(s)',fontsize=20)
plt.legend()
plt.grid(True)
```

```python
plt.figure(figsize=(20,10))
s=np.linspace(0,4*np.pi,100)
y=np.cos(s)
x=np.sin(s)
plt.bar(s,x,width=1,edgecolor='red')
plt.bar(s,y,width=1,color='yellow')
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('_mpl-gallery')

# make data:
np.random.seed(3)
x = 0.5 + np.arange(8)
y = np.random.uniform(2, 7, len(x))

# plot
fig, ax = plt.subplots()

ax.bar(x, y, width=1, edgecolor="white", linewidth=0.7)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```
