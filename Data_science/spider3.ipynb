```python
import requests
from bs4 import BeautifulSoup
```

```python
def load_soup(url):
    page=requests.get(url)
    if page.status_code==200:
        return BeautifulSoup(page.text)
```

```python
url='https://www.makemytrip.com/railways/lucknow-allahabad-train-tickets.html'
soup=load_soup(url)
if soup is not None :
    print('Data loaded')
else:
    print('Data could not be loaded')
```
